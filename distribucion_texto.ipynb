## Tarea 2

Programa que tome un texto de entrada, elimine signos de puntuación y dígitos, separe en palabras el documento, regrese el conteo de vocabulario y calcule la distribución del vocabulario en el texto dado

## Limpiar el texto

```python
import string
import pandas as pd

filename = input('Ingrese el nombre del archivo: ')
texto = open(filename, "r").read()
texto_sin_punt = texto.translate(str.maketrans('', '', string.punctuation))
texto_con_digitos = texto_sin_punt.replace('\n', '').replace('\t', '')
texto_limpio = ''.join([i for i in texto_con_digitos if not i.isdigit()])
lista_palabras = texto_limpio.split(' ')
lista_palabras
```


## Conteo de vocabulario

```python
conteo = {}
for palabra in lista_palabras:
  if palabra in conteo.keys():
    conteo[palabra] += 1
  else:
    conteo[palabra] = 1
conteo_df = pd.DataFrame(conteo, index=['Conteo']).transpose()
conteo_df.to_csv('conteo.csv')
conteo_df
```

## Distribución

```python
total = len(lista_palabras)
distribucion = {}
for palabra in conteo.keys():
  distribucion[palabra] = conteo[palabra] / total
distribucion_df = pd.DataFrame(distribucion, index=['Distribucion']).transpose()
distribucion_df.to_csv('distribucion.csv')
distribucion_df
```
